## Data preprocessing

steps needed:
- mask creation (jules function)
- place mask and image in good paths 
- create the configuration 
- run the diffusion model

```python
import fabio
import os
import numpy as np

def process(name):
    files = os.listdir(name)
    with open(os.path.join(name, "mask"),"rb") as f:
    	mask = np.load(f)
    if not os.path.exists(os.path.join(name, "masked_results")):
        os.makedirs(os.path.join(name, "masked_results"))
    for file in files:
        _, ext = os.path.splitext(file)
        if ext == ".tif":
            mask_image(name, file, mask)

def mask_image(name, file, mask):
    fullname = os.path.join(name, file)
    print(fullname)
    new_file, _ = os.path.splitext(file)
    new_file += '_masked.tif'
    result_file = os.path.join(name, "masked_results", new_file)
    img = fabio.open(fullname).data
    img = img[img.shape[0]-mask.shape[0]:,img.shape[1]-mask.shape[1]:]
    img[mask > 0] = -1
    fabio.tifimage.tifimage(data=img).write(result_file)
    
    
def create_mask(fullname):
    print(fullname)
    img = fabio.open(fullname).data
    mask = np.zeros(img.shape)
    mask[img < 0] = 1
    filename = os.path.join(os.path.split(fullname)[0], "qf_results", "mask")
    print(os.path.split(fullname)[0])
    with open(filename,"wb") as f:
        np.save(f, mask)

if __name__ == "__main__":
    create_mask("/home/jules/data/2021_0703_Gong/P1/P1_F1_tet_1_100_005_0001.tif")
    process("/home/jules/data/2021_0703_Gong/P1/qf_results")
```

## tiff to png conversion of input image

```python
from PIL import Image
import os

def tiff_to_png(input_path, output_path):
    try:
        with Image.open(input_path) as img:
            img.save(output_path, format="PNG", compress_level=0)
        print("Conversion successful.")
    except Exception as e:
        print(f"Error during conversion: {e}")

input_tiff_file = "/home/tristan/Documents/data/2021_0703_Gong/P1/P1_F1_tet_1_100_005_0002.tif"
output_png_file = "/home/tristan/Documents/data/2021_0703_Gong/P1/test_img/test_image.png"

tiff_to_png(input_tiff_file, output_png_file)
```

```python
def png_to_tiff(input_path, output_path):
    try:
        with Image.open(input_path) as img:
            img.save(output_path, format="TIFF", compress_level=0)
        print("Conversion successful.")
    except Exception as e:
        print(f"Error during conversion: {e}")

input_png_file = "/home/tristan/Documents/data/2021_0703_Gong/P1/test_img/test_image.png"
output_tiff_file = "/home/tristan/Documents/data/2021_0703_Gong/P1/test_img/test_image.tif"

png_to_tiff(input_png_file, output_tiff_file)
```

```python
# let's crop the tif image to 256x256, then convert to png, then convert it back to tif
from PIL import Image
import os

def crop_tif(input_path, output_path):
    """
    Crop the tif image to 256x256
    Params: fullname: the path to the image

    """
    img = Image.open(input_path)
    img = img.resize((256, 256))
    img.save(output_path, format="TIFF")
    print("Conversion successful.")


def create_gt_keep_mask(input_path, output_path):
    """
    Create a mask for the image, where the black pixels are 0 and the white pixels are 255
    Params: fullname: the path to the image

    """
    img = Image.open(input_path)
    img_matrix = np.array(img)
    matrix = np.zeros(img_matrix.shape)
    matrix[img_matrix < 0] = 0
    matrix[img_matrix >= 0] = 255
    # save the mask as a png
    mask = Image.fromarray(matrix)
    # display the mask:
    mask.show()
    # to prevent the error "OSError: cannot write mode F as TIFF"
    mask = mask.convert('RGB')
    mask.save(output_path, format="PNG")
    print("Conversion successful.")

# let's create the function that will apply the mask to the image
def apply_mask(input_path, mask_path, output_path):
    """
    Apply the mask to the image
    Params: fullname: the path to the image
            mask_path: the path to the mask
    """
    img = Image.open(input_path)
    mask = Image.open(mask_path)
    img_matrix = np.array(img)
    mask_matrix = np.array(mask)
    img_matrix[mask_matrix == 0] = 0
    img = Image.fromarray(img_matrix)
    
    # save the masked image 
    img.save(output_path, format="PNG", compress_level=0)


```

```python
from io import BytesIO

import torch

import PIL
import requests
from diffusers_modified.src.diffusers.pipelines.repaint.pipeline_repaint import RePaintPipeline
#from diffusers_modified.src.diffusers import RePaintScheduler
from diffusers import RePaintScheduler
import numpy as np

# let's create the mask for the image using create_gt_keep_mask
original_image_path = "/home/tristan/Documents/data/2021_0703_Gong/P1/P1_F1_tet_1_100_005_0002.tif"
create_gt_keep_mask(original_image_path, "/home/tristan/Documents/data/2021_0703_Gong/P1/test_masks/mask.png")
# create a png version of the image
tiff_to_png(original_image_path, "/home/tristan/Documents/data/2021_0703_Gong/P1/test_img/test_image.png")
path_to_image = "/home/tristan/Documents/data/2021_0703_Gong/P1/test_img/test_image.png"




def download_image(url):
    response = requests.get(url)
    return PIL.Image.open(BytesIO(response.content)).convert("RGB")


original_image_path = "/home/tristan/Documents/data/2021_0703_Gong/P1/P1_F1_tet_1_100_005_0002.tif"
mask_path = "/home/tristan/Documents/data/2021_0703_Gong/P1/test_masks/mask.png"

# Load the original image and the mask as PIL images
original_image = Image.open(path_to_image).resize((256, 256))
mask_image = Image.open(mask_path).resize((256, 256))
print(original_image.size)
# print color depth
print(original_image.mode)

# Load the RePaint scheduler and pipeline based on a pretrained DDPM model
scheduler = RePaintScheduler.from_pretrained("google/ddpm-ema-celebahq-256")
pipe = RePaintPipeline.from_pretrained("google/ddpm-ema-celebahq-256", scheduler=scheduler)
pipe = pipe.to("cuda")

generator = torch.Generator(device="cuda").manual_seed(0)
output = pipe(
    image=original_image,
    mask_image=mask_image,
    num_inference_steps=250,
    eta=0.0,
    jump_length=10,
    jump_n_sample=10,
    generator=generator,
)
inpainted_image = output.images[0]
```
